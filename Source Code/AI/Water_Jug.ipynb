```python
def fill1(m,jug2):
    jug1=m
    return jug1,jug2
def fill2(jug1,n):
    jug2=n
    return jug1,jug2
def empty1(jug2):
    jug1=0
    return jug1,jug2
def empty2(jug1):
    jug2=0
    return jug1,jug2
def jug1To2(m,n,jug1,jug2):
    space2=n-jug2
    space1=m-jug1
    if(space2>=jug1):
        jug2=jug2+jug1
        jug1=0
    elif(space2<jug1):
        jug1=jug1-space2
        jug2=n
    return jug1,jug2
def jug2To1(m,n,jug1,jug2):
    space1=m-jug1
    space2=n-jug2
    if(space1>=jug2):
        jug1=jug1+jug2
        jug2=0
    elif(space1<jug2):
        jug2=jug2-space1
        jug1=m
    return jug1,jug2
def check(d,jug1,jug2):
    global count 
    if count == 0:
        print("Initial State:", (jug1,jug2))
        count = count + 1
    elif count>0:
        print("\nStep" , count , ":" , (jug1,jug2))
        count = count + 1
    if((jug1, jug2) == (d,0) or (jug1,jug2)==(0,d)):
        print("\nGoal State Reached")
        return 1
    
m = int(input("Enter the capacity of Jug 1: "))
n = int(input("Enter the capacity of Jug 2: "))
goal = int(input("Enter the goal state: "))

jug1 = 0
jug2 = 0
count = 0

while (check(goal, jug1, jug2)!=1):
    c = int(input("\nSelect the operator to be performed: \n 1. Fill Jug 1\n 2. Fill Jug 2\n 3. Empty Jug 1\n 4. Empty Jug 2\n 5. Pour from Jug 1 to Jug 2\n 6. Pour from Jug 2 to Jug 1\n Choice: "))
    if c==1:
        jug1 , jug2 = fill1(m, jug2)
    elif c==2:
        jug1 , jug2 = fill2(jug1, n)
    elif c==3:
        jug1 , jug2 = empty1(jug2)
    elif c==4:
        jug1 , jug2 = empty2(jug1)
    elif c==5:
        jug1 , jug2 = jug1To2(m,n,jug1,jug2)
    elif c==6:
        jug1 , jug2 = jug2To1(m,n,jug1,jug2)
        
```
