```python
board=["-","-","-",
       "-","-","-",
       "-","-","-"]

currentPlayer='X'
winner=None
gameR=True
win = 0

def printboard(board):
    print(board[0]+ " | "+board[1]+ " | "+board[2])
    print("----------")
    print(board[3]+ " | "+board[4]+ " | "+board[5])
    print("----------")
    print(board[6]+ " | "+board[7]+ " | "+board[8])
    

# printboard(board)

def playerInput(board):
    pi=int(input("Enter a number between 1-9: "))
    if pi>=1 and pi<=9 and board[pi-1]=="-":
        board[pi-1]=currentPlayer
    else:
        print("This spot is already taken try another\n")
        playerInput(board)

def checkH(board):
    global winner
    if board[0]==board[1]==board[2]and board[1]!='-':
        winner=board[0]
        return True
    elif board[3]==board[4]==board[5]and board[3]!='-':
        winner=board[3]
        return True
    elif board[6]==board[7]==board[8]and board[6]!='-':
        winner=board[6]
        return True

def checkR(board):
    global winner
    if board[0]==board[3]==board[6] and board[0]!='-':
        winner=board[0]
        return True
    elif board[1]==board[4]==board[7] and board[1]!='-':
        winner=board[1]
        return True
    elif board[2]==board[5]==board[8] and board[2]!='-':
        winner=board[2]
        return True
    
def digonal(board):
    global winner
    if board[0]==board[4]==board[8] and board[0]!='-':
        winner=board[0]
        return True
    elif board[2]==board[4]==board[6] and board[6]!='-':
        winner=board[2]
        return True

def checkTie(board):
    global gameR
    if '-' not in board:
        printboard(board)
        print('It is a Tie !!!! \n Better Luck next time....')
        gameR= False


def Player():
    global currentPlayer
    if currentPlayer=='X':
        currentPlayer='O'
    else:
        currentPlayer='X'

def checkWin():
    if checkH(board) or checkR(board) or digonal(board):
        printboard(board) 
        print(f'The winner is {winner}')
        global win
        win =1

while gameR:
    printboard(board)
    playerInput(board)
    checkWin()
    if win == 1:
        break
    
    checkTie(board)
    Player()
```
